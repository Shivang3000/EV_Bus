```python
import pandas as pd
import matplotlib.pyplot as plt

# Define file path
file_path = "/content/drive/MyDrive/EV Data/vehicle_category_sales.csv"

# Read CSV file
df = pd.read_csv(file_path)

# Extract relevant data (2023-24 column)
categories = df["Category"]
sales_2023_24 = df["2023-24"]

# Plot pie chart
plt.figure(figsize=(5, 5))
plt.pie(sales_2023_24, labels=categories, autopct='%1.1f%%', startangle=140, colors=["lightblue", "lightgreen", "orange", "red"])
plt.title("Vehicle Sales Distribution (2023-24)")
plt.axis("equal")  # Ensures pie is drawn as a circle
plt.show()
```

```python
import pandas as pd

# Load the datasets
automobiles_data = pd.read_csv('/content/drive/MyDrive/EV Data/automobiles_produced_sold.csv')
emission_data = pd.read_csv('/content/drive/MyDrive/EV Data/emission_different_vehicles.csv')
emission_factors_data = pd.read_csv('/content/drive/MyDrive/EV Data/emission_factors.csv')
fleet_data = pd.read_csv('/content/drive/MyDrive/EV Data/fleet_data.csv')
tata_sales_data = pd.read_csv('/content/drive/MyDrive/EV Data/tata_motors_sales_commercial_vs_passenger.csv')
vehicle_sales_data = pd.read_csv('/content/drive/MyDrive/EV Data/vehicle_category_sales.csv')
```

```python
# Plotting the number of automobiles produced vs. sold
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(automobiles_data['Fiscal Year'], automobiles_data['Automobiles Produced (millions)'], label='Produced (millions)', marker='o')
plt.plot(automobiles_data['Fiscal Year'], automobiles_data['Automobiles Sold (millions)'], label='Sold (millions)', marker='o')
plt.title("Automobiles Produced vs Sold (in millions)")
plt.xlabel("Fiscal Year")
plt.ylabel("Amount (millions)")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Plot CO2 emissions for each category
plt.figure(figsize=(12,6))
plt.bar(emission_data['Categories'], emission_data['CO2 (Gg)'], color='skyblue')
plt.title("CO2 Emissions by Vehicle Category (in Gg)")
plt.xlabel("Vehicle Category")
plt.ylabel("CO2 Emissions (Gg)")
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
# Plot CO2 emission factors per km for each vehicle type
plt.figure(figsize=(12,6))
plt.bar(emission_factors_data['Vehicle Type'], emission_factors_data['CO2 (g/km)'], color='lightgreen')
plt.title("CO2 Emissions per Kilometer for Different Vehicle Types (in g/km)")
plt.xlabel("Vehicle Type")
plt.ylabel("CO2 Emissions (g/km)")
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
# Example: Calculate total CO2 emissions for each vehicle type based on emission factor and number of vehicles produced
emission_data['Total CO2 Emissions (Gg)'] = emission_data['CO2 (Gg)'] * (automobiles_data['Automobiles Produced (millions)'].sum() / len(automobiles_data))
print(emission_data[['Categories', 'Total CO2 Emissions (Gg)']])
```
